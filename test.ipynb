```python
%pip install --quiet pint
%pip install --quiet open_atmos_jupyter_utils

import pint
import scipy
si = pint.UnitRegistry()
import matplotlib as plt
from matplotlib import pyplot
si.setup_matplotlib()
import numpy as np
from open_atmos_jupyter_utils import show_plot

R_str = ( scipy.constants.R * si.J/si.K/si.mol )
M = 28.966 * si.g/si.mol
p = 1013*si.hPa
```

```python
print(2*si.cd)
```

```python
def rho(p, T):
    return p/R_str/T * M
```

```python
air_rho = rho(p= 1013*si.hPa, T= (273+20)*si.K).to_base_units()
print(air_rho)
```

```python
print(f'{air_rho.to(si.lbs/si.inch**3)=:.3g}')
```

```python
pyplot.plot( T:= np.linspace(200, 300)*si.K, rho(p, T).to_base_units(), color= 'purple', label= 'rho')
pyplot.gca().xaxis.set_units(si.celsius)
pyplot.legend()
show_plot()
```
